5. Operadores

```python
# 5.1 Operadores Aritméticos
# Se tiene: +,-, *, /,%, **,//
```

```python
numero1 = 10
numero2 = 3
suma = numero1 + numero2
resta = numero1 - numero2
multiplicación = numero1 * numero2
división = numero1 / numero2
residuo = numero1 % numero2
potencia = numero1 ** numero2
cociente = numero1 // numero2
print(suma)
print(resta)
print(multiplicación)
print(división)
print(residuo)
print(potencia)
print(cociente)
```

```python
# 5.2 Operadores de asignación
# Se tiene: =,+=, -=, *=, /=, **=
numero = 5
numero = numero + 3
print(numero)
```

```python
# La operación anterior tambien se puede expresar de la siguiente manera:
numero = 5
numero += 3
print(numero)
```

```python
# Ejemplo:
numero = 5
numero = numero ** 3
print(numero)
```

```python
# La operación anterior tambien se puede expresar de la siguiente manera:
numero = 5
numero **= 3
print(numero)
```

```python
# 5.3 Operadores de comparación
# Se tiene: ==, !=, <, >,<=,>=,
```

```python
# Ejemplo de igualdad ==
numero1 = 5
numero2 = 2
numero1 == numero2
```

```python
# Con las cadenas funciona igual
nombre1 = "Frank"
nombre2 = "Frank"
nombre1 == nombre2
```

```python
# 5.4 Operadores Lógicos
# Se tiene:  and, or, not
numero1 = 5
numero2 = 6
numero3 = 7
numero4 = 8
```

```python
numero1 < numero2
```

```python
# el Not invierte el resultado
not(numero1<numero2)
```

```python
numero1<numero2 and numero3<numero4
```

```python
# Tambien puedes utilizar paraentesis para ordenar tu operación
(numero1<numero2 and numero3<numero4) or (numero3>numero4)
```

```python
# 5.5 Operadores de Pertenencia
# Se tiene: in, not in
```

```python
# Ejemplo:
aula = ["Juan", "Daniel", "Ruben"]
alumno1 = "Diego"
alumno2 = "Juan"
```

```python
# ¿Diego está en aula?
alumno1 in aula
```

```python
# ¿ Juan está en aula?
alumno2 in aula
```

```python
# ¿ Diego no está en aula?
alumno1 not in aula
```

```python
# ¿ Juan no está en aula?
alumno2 not in aula
```

```python
# 5.6 Operadores de Identidad
# Se tiene: is, is not
```

```python
aula1 = ["Juan", "Daniel", "Ruben"]
aula2 = ["Juan", "Daniel", "Ruben"] 
#  A continuación vamos a indicar que aula3 va a compartir el mismo objeto que aula1
aula3 = aula1
```

```python
# ¿aula3 comparte el mismo objeto que aula1?
aula3 is aula1
```

```python
# aula3 comparte el mismo objeto que aula2?
aula3  is aula2
```

```python
# aula2 comparte el mismo objeto que aula1?
aula2 is aula1
```
